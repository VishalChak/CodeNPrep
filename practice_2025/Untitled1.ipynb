```python
arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
```

```python
def max_sum_arr(arr):
    max_sum = 0
    curr_sum = 0
    for i in range(len(arr)):
        curr_sum = max(arr[i] , (curr_sum+ arr[i]))
        max_sum = max(curr_sum, max_sum)
    return max_sum
```

```python
max_sum_arr(arr)
```
