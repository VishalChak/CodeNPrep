```python
### Continus maxsum sub array 
```

```python
arr= [1,2,3,4-5,6,7,8,9]
```

```python
def max_sub_arr(arr):
    max_sum = 0
    curr_sum = 0
    for i in range(len(arr)):
        curr_sum = max(arr[i], curr_sum + arr[i])
        max_sum = max(max_sum, curr_sum)
    return max_sum
```

```python
max_sub_arr(arr)
```
