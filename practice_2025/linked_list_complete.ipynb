```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
    def insert(self, data):
        node = Node(data)
        if self.head == None:
            self.head = node
        else:
            temp = self.head
            while temp.next != None:
                temp = temp.next
            temp.next = node
        print(f"Inserted Node{data}")
    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end = ",")
            temp = temp.next
        print()
    def delete_node(self, data):
        temp = self.head
        if temp and temp.data == data:
            self.head = temp.next
            print(f"data Deleted {data}")
        elif temp.next != None:
            prev = temp
            temp = temp.next
            found = False
            while temp:
                if temp.data  == data:
                    prev.next = temp.next
                    print(f"Data Deleted {data}")
                    found = True
                    break
                prev = temp
                temp = temp.next
            if not found:
                print(f"Data not found {data}")
        else:
            print(f"Data Not found {data}")
    def get_len(self):
        len_ = 0
        temp = self.head
        while temp:
            len_ +=1
            temp= temp.next
        print(f"Len of List {len_}")
        return len_
    def reverse(self):
        len_ = self.get_len()
        print(len_)
        if len_ > 1:
            prev_ = None
            curr_ = self.head
            next_ = curr_.next
            while curr_:
                curr_.next = prev_
                prev_ = curr_
                curr_ = next_
                if curr_:
                    next_ = curr_.next
            self.head = prev_
        
    
```

```python
ll = LinkedList()

for i in range(1):
    ll.insert(i)
ll.print_list()
ll.insert(23)
ll.print_list()
ll.delete_node(2)
ll.delete_node(0)
ll.print_list()

ll.reverse()
ll.print_list()
```
