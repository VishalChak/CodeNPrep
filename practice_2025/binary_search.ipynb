```python
### Using recussion

def binary_search(arr, val, low, high):
    if low > high:
        return None
    mid = (high + low) // 2
    
    if arr[mid] == val:
        return mid
    elif val > arr[mid]:
        return binary_search(arr, val , mid+1, high)
    else:
        return binary_search(arr, val, low, mid -1 )

def b_search_loop(arr, val):
    low = 0
    high = len(arr)-1
    while low < high:
        mid = (low+ high ) //2
        if val == arr[mid]:
            return mid
        elif val > mid:
            low = mid+1
        else:
            high = mid -1
    
```

```python
arr = [val for val in range(10)]
```

```python
ans = binary_search(arr,4 , 0, len(arr)-1)
ans_1 = b_search_loop(arr, 5)
print(ans)
print(ans_1)
```
